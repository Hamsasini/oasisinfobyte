```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import register_matplotlib_converters

# Ensure matplotlib converters are registered
register_matplotlib_converters()

# Step 1: Load the Dataset
file_path = r"C:\Users\LENOVO\Desktop\oasis1 dataset1.csv"  
data = pd.read_csv(file_path)

# Step 2: Inspect the Dataset
print("First few rows of the dataset:")
print(data.head())
print("\nColumns in the dataset:")
print(data.columns)

# Step 3: Remove Duplicates
data_cleaned = data.drop_duplicates()
print("\nDuplicates removed.")

# Step 4: Convert Date Column
data_cleaned['invoice_date'] = pd.to_datetime(data_cleaned['invoice_date'], dayfirst=True, errors='coerce')
print("\nDate conversion done. Sample data:")
print(data_cleaned.head())

# Ensure there are no missing values after conversion
data_cleaned = data_cleaned.dropna(subset=['invoice_date'])
print(f"\nRows after dropping NA in 'invoice_date': {data_cleaned.shape[0]}")

# Step 5: Aggregate Sales Data
# Assuming 'quantity' represents sales, we aggregate sales per day
data_cleaned.set_index('invoice_date', inplace=True)
daily_sales = data_cleaned.resample('D').sum()['quantity']

# Ensure there are no missing values after resampling
daily_sales = daily_sales.dropna()
print("\nDaily sales calculated. Sample data:")
print(daily_sales.head())

# Step 6: Plot the Time Series
plt.figure(figsize=(12, 6))
plt.plot(daily_sales, label='Daily Sales')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Quantity')
plt.legend()
plt.show()

# Step 7: Decompose the Time Series
decomposition = seasonal_decompose(daily_sales, model='additive')
decomposition.plot()
plt.show()

# Step 8: Forecasting with ARIMA
# Differencing to make the series stationary
daily_sales_diff = daily_sales.diff().dropna()
print("\nDifferenced series calculated. Sample data:")
print(daily_sales_diff.head())

# Fit the ARIMA model
model = ARIMA(daily_sales_diff, order=(1, 1, 1))
fit = model.fit()

# Forecasting the next 30 days
forecast = fit.forecast(steps=30)
forecast_cumsum = forecast.cumsum()
forecast_values = daily_sales[-1] + forecast_cumsum
print("\nForecast calculated. Sample data:")
print(forecast_values.head())

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(daily_sales, label='Observed')
plt.plot(forecast_values, label='Forecast', linestyle='--')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales Quantity')
plt.legend()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\LENOVO\Desktop\oasis1 dataset1.csv"
data = pd.read_csv(file_path)

# Step 1: Inspect the Dataset
print("First few rows of the dataset:")
print(data.head())
print("\nColumns in the dataset:")
print(data.columns)

# Step 2: Remove Duplicates
data_cleaned = data.drop_duplicates()
print("\nDuplicates removed.")

# Step 3: Handle Missing Values
print("\nMissing values before cleaning:")
print(data_cleaned.isnull().sum())
data_cleaned = data_cleaned.dropna()
print("\nMissing values after cleaning:")
print(data_cleaned.isnull().sum())

# Step 4: Convert Date Column
data_cleaned['invoice_date'] = pd.to_datetime(data_cleaned['invoice_date'], dayfirst=True, errors='coerce')

# Ensure there are no missing values after conversion
data_cleaned = data_cleaned.dropna(subset=['invoice_date'])
print(f"\nRows after dropping NA in 'invoice_date': {data_cleaned.shape[0]}")

# Step 5: Analyze Customer Demographics
# Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender Distribution
plt.figure(figsize=(6, 6))
data_cleaned['gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution of Customers')
plt.ylabel('')
plt.show()

# Step 6: Analyze Purchasing Behavior
# Average Spending per Customer
data_cleaned['total_spent'] = data_cleaned['quantity'] * data_cleaned['price']
avg_spending = data_cleaned.groupby('customer_id')['total_spent'].mean()

plt.figure(figsize=(10, 6))
sns.histplot(avg_spending, bins=20, kde=True)
plt.title('Average Spending per Customer')
plt.xlabel('Average Spending')
plt.ylabel('Frequency')
plt.show()

```
