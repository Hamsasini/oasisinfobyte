```python
import pandas as pd
from scipy import stats

# Load the Excel file
file_path = r"C:\Users\LENOVO\Desktop\shopping trends.xlsx"
df = pd.read_excel(file_path)

# Display column names to verify
print("Column names in the dataset:")
print(df.columns)

# Updated column names based on your file
required_columns = [ 'Customer ID', 'Age', 'Purchase Amount (USD)']

# Check for required columns
for column in required_columns:
    if column not in df.columns:
        raise KeyError(f"Column '{column}' is missing from the dataset.")

# Remove duplicates and handle missing values
df_cleaned = df.drop_duplicates()
print(df_cleaned)

# Calculate key metrics

# Average purchase value
average_purchase_value = df_cleaned['Purchase Amount (USD)'].mean()

# Frequency of purchases per customer
purchase_frequency = df_cleaned.groupby('Customer ID').size().mean()

# Descriptive statistics for age
mean_age = df_cleaned['Age'].mean()
median_age = df_cleaned['Age'].median()
mode_age = stats.mode(df_cleaned['Age'])[0]
std_dev_age = df_cleaned['Age'].std()

# Descriptive statistics for Purchase Amount(USD)
mean_total_spent = df_cleaned['Purchase Amount (USD)'].mean()
median_total_spent = df_cleaned['Purchase Amount (USD)'].median()
mode_total_spent = stats.mode(df_cleaned['Purchase Amount (USD)'])[0]
std_dev_total_spent = df_cleaned['Purchase Amount (USD)'].std()

# Print the results
print(f"Average Purchase Value: {average_purchase_value}")
print(f"Average Frequency of Purchases per Customer: {purchase_frequency}")

print(f"Mean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print(f"Standard Deviation of Age: {std_dev_age}")

print(f"Mean Purchase Amount: {mean_total_spent}")
print(f"Median Purchase Amount: {median_total_spent}")
print(f"Mode Purchase Amount: {mode_total_spent}")
print(f"Standard Deviation of Purchase Amount: {std_dev_total_spent}")



```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = r"C:\Users\LENOVO\Desktop\shopping trends.xlsx"
df = pd.read_excel(file_path)

# Check the first few rows of the dataframe to ensure it's loaded correctly
print(df.head())

# Select the features for clustering
X = df[['Age', 'Purchase Amount (USD)']]

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.scatter(df['Age'], df['Purchase Amount (USD)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.title('Customer Segmentation using K-means Clustering')
plt.show()

# Print the resulting DataFrame with cluster labels
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = r"C:\Users\LENOVO\Desktop\shopping trends.xlsx"
df = pd.read_excel(file_path)

# Display column names to verify
print("Column names in the dataset:")
print(df.columns)

# Updated column names based on your file
required_columns = ['Customer ID', 'Age', 'Purchase Amount (USD)']

# Check for required columns
for column in required_columns:
    if column not in df.columns:
        raise KeyError(f"Column '{column}' is missing from the dataset.")

# Remove duplicates and handle missing values
df_cleaned = df.drop_duplicates()


# Set the plotting style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_cleaned, x='Age', y='Purchase Amount (USD)', palette='viridis', s=100)
plt.title('Scatter Plot: Age vs. Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.show()

# Bar Chart: Average Purchase Amount by Age Group
df_cleaned['Age Group'] = pd.cut(df_cleaned['Age'], bins=[0, 20, 30, 40, 50, 60, 100], labels=['0-20', '21-30', '31-40', '41-50', '51-60', '60+'])
avg_purchase_by_age_group = df_cleaned.groupby('Age Group')['Purchase Amount (USD)'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=avg_purchase_by_age_group, x='Age Group', y='Purchase Amount (USD)', palette='viridis')
plt.title('Average Purchase Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()

# 3. Histogram: Distribution of Age
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['Age'], kde=True, bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 4. Box Plot: Purchase Amount Distribution by Age Group
plt.figure(figsize=(14, 7))
sns.boxplot(data=df_cleaned, x='Age Group', y='Purchase Amount (USD)', palette='viridis')
plt.title('Box Plot: Purchase Amount Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```
